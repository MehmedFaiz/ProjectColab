```python
import pandas as pd
from category_encoders import TargetEncoder

data = {"produk": ["laptop","Smartphone","Headphone","TV LED","Kulkas"]}
df = pd.DataFrame(data)

te = TargetEncoder(cols=["produk"])
df_encoded = te.fit_transform(df, y=df["produk"])
print(df_encoded)
df["product_encoded"] = te.transform(df["produk"])
print(df)
```

```python
df_encoded = pd.get_dummies(df, columns=["produk"])
print(df_encoded)
```
