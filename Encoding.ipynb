```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {"produk": ["laptop","Smartphone","Headphone","TV LED","Kulkas"]}
df = pd.DataFrame(data)

le = LabelEncoder()
df["product_encoded"] = le.fit_transform(df["produk"])
print(df)
```
