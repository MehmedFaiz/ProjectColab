```python
import pandas as pd

file_path = "sales_data.csv"
df = pd.read_csv(file_path)
print(df.head())
print(df.info())
print(df.describe())
```

After uploading, the file will be available in the Colab environment. You can then use its name directly, like in your original code: `file_path = "sales_data.csv"`.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['tanggal'] = pd.to_datetime(df['tanggal'])

df_daily = df.groupby(df['tanggal'].dt.date)['total_penjualan'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=df_daily['tanggal'].astype(str), y=df_daily['total_penjualan'], marker='o')
plt.xticks(rotation=45)
plt.xlabel("Tanggal")
plt.ylabel("Total Penjualan")
plt.title("Tren Penjualan Harian")
plt.show()
```
